```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: make plots look nicer
sns.set(style="whitegrid", palette="pastel", font_scale=1.1)
plt.rcParams['figure.figsize'] = (10,6)

# Load cleaned dataset
df = pd.read_csv("cleaned_students.csv")
df.head(5)
```

```python
# Histogram of Final_Score
plt.figure()
sns.histplot(df["Final_Score"], bins=15, kde=True, color='skyblue')
plt.title("Distribution of Final Scores")
plt.xlabel("Final Score")
plt.ylabel("Number of Students")
plt.show()
```

**Insight:**
- Most students scored between 60-80 in the final exam.
- Very few students scored below 40 or above 95.
- The distribution is slightly right-skewed.
---

```python
# Bar plot of average final score per department
plt.figure()
dept_avg = df.groupby("Department")["Final_Score"].mean().sort_values()
sns.barplot(x=dept_avg.index, y=dept_avg.values, palette="pastel")
plt.title("Average Final Score by Department")
plt.xlabel("Department")
plt.ylabel("Average Final Score")
plt.xticks(rotation=45)
plt.show()
```

**Insight:**
- Some departments have consistently higher average scores.
- This could indicate differences in curriculum or teaching methods.
---

```python
# Scatter plot: Study Hours vs Final Score
plt.figure()
sns.scatterplot(x="Study_Hours_per_Week", y="Final_Score", data=df, hue="Gender", palette="bright")
plt.title("Study Hours per Week vs Final Score")
plt.xlabel("Study Hours per Week")
plt.ylabel("Final Score")
plt.show()
```

**Insight:**
- Students who study more tend to get higher final scores.
- There is some variation; study hours alone dont fully predict performance.
- Gender differences appear minimal.
- ---



```python
# Box plot: Attendance (%) by Grade
plt.figure()
sns.boxplot(x="Grade", y="Attendance (%)", data=df, palette="Set2", hue=None)
plt.title("Attendance (%) by Grade")
plt.xlabel("Grade")
plt.ylabel("Attendance (%)")
plt.show()
```

**Insight:**
-  Students with higher grades generally have higher attendance.
- Outliers exist: some students with low attendance still achieved good grades.
- ---


```python
# Count plot: Students by Gender
plt.figure()
sns.countplot(x="Gender", data=df, palette="pastel")
plt.title("Number of Students by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

 **Insight:**
- There are slightly more male students than female students in this dataset.
- This may affect overall performance statistics if analyzed by gender.
